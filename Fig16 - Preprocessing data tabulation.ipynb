### For setting up latent space analysis

```python
%load_ext autoreload
%autoreload 2

import json, os

import numpy as np

from Experiment import Experiment
```

```python
%%time

# set results folder
rf = "results/Appendix-100Experiments"

# List for saving results
latents = []

# Iterate through all the model results, and save them!
for modelnum in range(1, 100):
    try:
        # Set the experiment name
        expt_name = "S0-{}".format(str(modelnum))
        # Load up the different experiments for analysis
        expt = Experiment(experiment_name=expt_name, data_file_prefix='S0-Oscillator', results_folder=rf)
        # Get latent vectors
        v, f = expt.encode_vf(index=10)
        latents.append((v, f))
        # Delete the Experiment object
        del(expt)
    except:
        continue
```

```python
v, f = latents[0]
```

```python
all_v = []
all_f = []
for (v,f) in latents:
    all_v.append(v)
    all_f.append(f)
    
all_v = np.asarray(all_v)
all_f = np.asarray(all_f)
```

```python
np.save('./results/Appendix-100Experiments/S0-all_v', all_v)
np.save('./results/Appendix-100Experiments/S0-all_f', all_f)
```
