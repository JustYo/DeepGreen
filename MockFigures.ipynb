```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

probs = ["2-NLSL", "2v-NLSL", "4-NLSL", "2D Poisson"]

n = 1000
means = [1e-4, 7e-4, 3.5e-4, 1e-3]
spreads = [3e-5, 1e-4, 7e-5, 2e-4]

data = []
for i in range(len(probs)):
    data.append(np.random.normal(means[i], spreads[i], n))

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.set_title(r'$\mathbf{U}(x)$ Prediction Error Given $\mathbf{F}(x)$')
ax.set_ylabel("Prediction Error")
ax.set_xlabel("Problem Type")
ax.boxplot(data, labels=probs)
#ax.ticklabel_format(axis='y', style='sci')
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.2e'))
plt.show()
```
