```python
expt_name = "NLSL_Experiment_05d"
model_names = ["model_{}".format(str(i)) for i in range(20)]

# Network architecture design
l = 20  # Latent space size

# Set up validation data
val_x = [data_val_u, data_val_f]
val_ys = [data_val_u, data_val_f, data_val_f, data_val_u]


for model_name in model_names:
    # Set model path and latent space size
    best_model_path = "./model_weights/{}/{}".format(expt_name, model_name)

    # Load the best model
    full_model = keras.models.load_model(best_model_path, 
                                         custom_objects={"NormalizedMeanSquaredError": NMSE})

    # Compute losses by evaluate method
    losses = full_model.evaluate(x=val_x, y=val_ys, batch_size=64, verbose=0)
    
    # print the results:
    print(expt_name)
    print(model_name)
    for loss in losses:
        print(loss)
    #print("--------------")
```
